#Data selection
This notebook selects takes the raw data as an input, drops the unused data, creates more features and splits the data into Train and Test datasets

```
from google.colab import drive
drive.mount('/content/gdrive')
```

```
import pandas as pd
import plotly.express as px
```

```
df_raw=pd.read_csv('/content/gdrive/My Drive/Data Science/Electric_motor_temperature/01_Data/01_Raw/pmsm_temperature_data.csv') #raw data frame, from https://www.kaggle.com/wkirgsn/electric-motor-temperature/download
df_raw=df_raw.drop(columns='torque')  #We drop the torque because as it is mentionned, it doesn't reprensent a reliable variable

df_raw.columns=['ambient', 'coolant', 'u_d', 'u_q', 'speed', 'i_d', 'i_q', 'pm','yoke', 'tooth', 'winding', 'profile_id']   #we rename the columns

df_test=df_raw[(df_raw.profile_id==20)|(df_raw.profile_id==65)].reset_index(drop=True)    #we keep id_20 and id_65 for testing
df_train=df_raw[(df_raw.profile_id!=20)&(df_raw.profile_id!=65)].reset_index(drop=True)    #the rest of the data for trainning
```

```
#saving the two files in the clean folder
df_test.to_csv('/content/gdrive/My Drive/Data Science/Electric_motor_temperature/01_Data/02_Clean/test.csv')
df_train.to_csv('/content/gdrive/My Drive/Data Science/Electric_motor_temperature/01_Data/02_Clean/train.csv')
```
